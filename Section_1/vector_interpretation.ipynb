## Vector dot product

```python
## many ways to compute the dot product
v1 = np.array([1, 2, 3, 4, 5, 6])
v2 = np.array([0, -4,2,-3, 6, 5])

# method 1
dp1 = sum(np.multiply(v1, v2))

# method 2
dp2 = np.dot (v1, v2)

# method 3
dp3 = np.matmul(v1, v2)

# method 4
dp4 = 0 #initialize

# Loop over elements
for i in range(0,len(v1)):
    #multiply corresponding element and sum
    dp4 = dp4 + v1[i]*v2[i]

print(dp1,dp2,dp3,dp4)
```

## Dot product properties

### distributive a(b+c) = ab + ac, aT(b+c) = aTb + aTc

### vector product is asscociative but dot product is not
### a(b . c) = (a. b)c
### aT(bTc) not equal (aTb)Tc



```python
# distibutive property
#create random vectors
n=10
a = np.random.randn(n)
b= np.random.rand(n)
c = np.random.randn(n)

#the two results
res1 = np.dot(a, (b+c))
res2 = np.dot(a,b) + np.dot(a,c)

#compare them
print([res1,res2])
```

```python
# Associative property
#create random vectors
n=5
a = np.random.randn(n)
b= np.random.rand(n)
c = np.random.randn(n)

#the two results
res1 = np.dot(a, np.dot(b,c))
res2 = np.dot(np.dot(a,b), c)

#compare them
print(res1)
print(res2)

### Special cases where associative property works!
# 1) one vector is the zeros vector
# 2) a==b==c
```

```python
# a vector
v1 = np.array([1, 2, 3, 4, 5, 6])

vl1 = np.sqrt(sum(np.multiply(v1,v1)))
vl2 = np.linalg.norm(v1)
print(vl1,vl2)
```

```python
# two vectors
v1 = np.array([16, -2, 4])
v2 = np.array([0.5, 2, -1])

# compute the angle radians between two vectors
angl = np.arccos(np.dot(v1,v2) / (np.linalg.norm(v1)*np.linalg.norm(v2)))
print(angl)

#draw them
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.plot([0, v1[0]],[0, v1[1]],[0, v1[2]],'b')
ax.plot([0, v2[0]],[0, v2[1]],[0,v2[2]],'r')

plt.axis((-6, 6, -6, 6))
plt.title('Angle between vectors: %s rad.' %angl)
plt.show()
```
